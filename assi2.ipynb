```python
df = df.drop(columns=["Title", "MiddleName", "Suffix", 
                      "AddressLine1", "AddressLine2", 
                      "PhoneNumber"])
df.head()
```

```python
data_types = {
    "Nominal-Discrete": [
        "City", "StateProvinceName", "CountryRegionName",
        "PostalCode", "Occupation", "Gender",
        "MaritalStatus", "HomeOwnerFlag"
    ],
    "Ordinal-Discrete": [
        "Education"
    ],
    "Interval-Continuous": [
        "BirthDate", "LastUpdated"
    ],
    "Ratio-Discrete": [
        "NumberCarsOwned", "NumberChildrenAtHome", "TotalChildren"
    ],
    "Ratio-Continuous": [
        "YearlyIncome"
    ]
}
print(data_types )
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# (a) Handling Null values
# Impute numeric with mean, categorical with mode
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

numeric_cols = df.select_dtypes(include=[np.number]).columns
categorical_cols = df.select_dtypes(exclude=[np.number]).columns

df[numeric_cols] = num_imputer.fit_transform(df[numeric_cols])
df[categorical_cols] = cat_imputer.fit_transform(df[categorical_cols])
```

```python
# (b) Normalization (Min-Max scaling for numeric columns)
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
```

```python
# (c) Discretization (binning continuous features)
# Example: bin YearlyIncome into 5 categories
df["YearlyIncome_Binned"] = pd.cut(df["YearlyIncome"], bins=5, labels=False)

# (d) Standardization
std_scaler = StandardScaler()
df[numeric_cols] = std_scaler.fit_transform(df[numeric_cols])

# (e) Binarization (One-Hot Encoding for categorical attributes)
df = pd.get_dummies(df, columns=categorical_cols)
```

```python
df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import jaccard
```

```python
row1, row2 = df.iloc[0].values.reshape(1, -1), df.iloc[1].values.reshape(1, -1)
cos_sim = cosine_similarity(row1, row2)[0][0]
jac_sim = 1 - jaccard(df.iloc[0].values, df.iloc[1].values)

# Simple Matching Coefficient (SMC)
smc = (df.iloc[0] == df.iloc[1]).sum() / len(df.columns)
```

```python
print("Cosine Similarity:", cos_sim)
print("Jaccard Similarity:", jac_sim)
print("Simple Matching Coefficient:", smc)
```
